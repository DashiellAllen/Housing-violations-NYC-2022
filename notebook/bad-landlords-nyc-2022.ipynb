```python
import pandas as pd
```

```python
dtypes = {
    "ViolationID":"str",
    "BuildingID":"str"
    
}


housing_data = pd.read_csv("../data/Housing_Maintenance_Code_Violations-2.csv",
                          dtype=dtypes
                          )
```
